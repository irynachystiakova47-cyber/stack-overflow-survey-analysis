Завдання 7. Географічний аналіз компенсації Python-розробників
-
- Інструмент, який потрібен для виконання цього завдання: Pandas (групування даних, агрегація, фільтрація)

- Опис завдання: Серед респондентів, що програмують на Python, проведи аналіз компенсації (ConvertedCompYearly) в розрізі країн.

- Для кожної країни обчисли:
> - Середню суму компенсації
> - Медіанну суму компенсації

```python
# Фільтрація респондентів, які програмують на Python
python_devs = df_p[df_p['LanguageHaveWorkedWith'].str.contains('Python', case = False, na = False)]
```

```python
# 7.1 Відбір даних з компенсацією та країною

# Залишаємо лише потрібні колонки
python_devs_comp = python_devs[['Country', 'ConvertedCompYearly']]

# Видаляємо рядки з пропущеною компенсацією
python_devs_comp = python_devs_comp.dropna(subset=['ConvertedCompYearly'])
```

```python
# 7.2 Групування по країнах та агрегація

country_compensation = (python_devs_comp
                        .groupby ('Country') ['ConvertedCompYearly']
                        .agg(
                            mean_compensation = 'mean',
                            median_compensation = 'median'
                        )
                        .reset_index()
                        .round()
                       )
country_compensation
```

```python
# 7.3 Сортуємо за середньою компенсацією (за спаданням)
country_compensation.sort_values(
    by='mean_compensation',
    ascending=False
).head(10)
```

ВИСНОВОК:

У межах завдання було проведено аналіз річної компенсації Python-розробників у розрізі країн. Дані було відфільтровано за використанням мови Python та наявністю значення компенсації. 
Для кожної країни обчислено середню та медіанну компенсацію. Результати показали значні відмінності між країнами, а також різницю між середнім і медіанним значеннями в окремих країнах (Viet Nam, Nigeria — дуже висока середня, але низька медіана), що може свідчити про вплив викидів або невелику кількість респондентів у вибірці.

```python
# Перевіряємо всі наявні вікові категорії
df_p['Age'].value_counts()
```

```python
# 9.1 Підготовка даних

# Залишаємо лише потрібні колонки
age_language = df_p [['Age','LanguageHaveWorkedWith']]

# Видаляємо респондентів без вікової категорії
age_language = age_language.dropna(subset = ['Age'])
```

```python
# 9.2 Додаємо ознаку використання Python

# Створюємо булеву колонку, яка показує, чи використовує респондент Python
age_language['use_python'] = age_language['LanguageHaveWorkedWith'].str.contains('Python', case = False, na = False)
```

```python
# 9.3 Групування по вікових категоріях

# Групуємо респондентів за віковими категоріями
python_by_age = (age_language
                 .groupby('Age')
                 .agg(
                     total_respondents = ('use_python', 'count'),
                     python_users = ('use_python', 'sum')
                 )
                )

# та рахуємо загальну кількість і кількість Python-розробників
python_by_age ['python_percentage'] = (
    (python_by_age['python_users']/python_by_age['total_respondents'])*100
).round()

python_by_age
```

У межах завдання було проаналізовано популярність мови програмування Python серед різних вікових категорій респондентів. 
Результати показали, що частка Python-розробників є відносно стабільною у більшості вікових груп (близько 37–40%), з найвищими значеннями у групах 18–24 та 45–54 роки. *Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
-
*(виконання необов'язкове, за бажанням)

- Інструмент, який потрібен для виконання цього завдання: Pandas (перцентилі, множинна фільтрація, підрахунок частоти)

- Опис завдання: Знайди респондентів, які:
> - Знаходяться у 75 перцентилі за компенсацією (тобто заробляють більше, ніж 75% респондентів)
> 
> - Працюють віддалено

Для цієї групи визнач найрозповсюдженіші індустрії.

```python
# 10.1 Підготовка даних і 75-й перцентиль

# Видаляємо респондентів без інформації про компенсацію
comp_data = df_p.dropna (subset = 'ConvertedCompYearly')

# Обчислюємо 75-й перцентиль компенсації
percentile_75 = comp_data['ConvertedCompYearly'].quantile(0.75)
percentile_75
```

```python
# 10.2 Фільтрація: працівників 1) з компенсацією вище 75-го перцентиля; 2) які працюють у повністю віддаленому форматі

high_paid_remote = comp_data [
    (comp_data['ConvertedCompYearly'] > percentile_75)&
    (comp_data['RemoteWork'] == 'Remote')
]
```

```python
# 10.3 Аналіз індустрій серед високооплачуваних працівників, які працюють віддалено

# Видаляємо респондентів без зазначеної індустрії
industry_data = high_paid_remote.dropna(subset=['Industry'])

# Підрахунок популярності індустрій
industry_counts = industry_data['Industry'].value_counts()

industry_counts
```

```python
# Перетворюємо у DataFrame
industry_table = industry_counts.reset_index()
industry_table.columns = ['Industry', 'Count']

industry_table
```

У межах завдання було проаналізовано індустрії серед респондентів, які входять до верхнього квартиля за рівнем річної компенсації та працюють у віддаленому форматі. 
Результати показали, що серед високооплачуваних віддалених працівників найбільш поширеною є індустрія Software Development, що значно випереджає інші сфери. Також серед популярних індустрій представлені "Fintech", "Healthcare" та "Internet, or Information Services", що свідчить про високий рівень цифровізації та попит на віддалену працю саме в цих галузях.
